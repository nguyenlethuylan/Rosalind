```python
# Overlap_Graphs
def read_fasta(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    seq_names = [line.strip()[1:] for line in lines if line.startswith(">")]
    sequences = []
    for i, line in enumerate(lines):
        if line.startswith(">"):
            sequence = ''
            for j in range(i + 1, len(lines)):
                if lines[j].startswith(">"):
                    break
                sequence += lines[j].strip()
            sequences.append(sequence)
    return {'names': seq_names, 'sequences': sequences}

def find_overlaps(strings, k=3):
    overlaps = {}
    for i in range(len(strings['sequences'])):
        for j in range(len(strings['sequences'])):
            if i != j:
                suffix = strings['sequences'][i][-k:]
                prefix = strings['sequences'][j][:k]
                if suffix == prefix:
                    overlaps[strings['names'][i] + ' ' + strings['names'][j]] = [strings['names'][i], strings['names'][j]]
    return overlaps

# Read DNA strings from file
dna_data = read_fasta("/home/lanlan/Downloads/rosalind_grph.txt")

# Find overlaps of length 3
overlap_graph = find_overlaps(dna_data, k=3)

# Print adjacency list with sample IDs
for edge, nodes in overlap_graph.items():
    print(' '.join(nodes))
```
